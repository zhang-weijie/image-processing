```python
from skimage import data
import matplotlib.pyplot as plt
im = data.astronaut()
#show the image, then call plt.show() immediately
plt.imshow(im)
plt.show()
```

```python
#print the rgb values of each pixel
print(im)
```

```python
lst = [x for x in range(10)]
lst[:3]
```

```python
# cut the region with height(10:180), width(120:310) and full color channel
im2 = im[10:180,120:310,:]
plt.imshow(im2)
plt.show()
```

```python
im3 = im.copy()
im3[10:180,120:310,:]=[255,0,0]
plt.imshow(im3)
plt.show()
```

```python
#convert colorful image to gray image
from skimage import color
im_gray = color.rgb2gray(im)
plt.imshow(im_gray)
plt.show()
```
